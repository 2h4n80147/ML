```python
import numpy as np
import matplotlib.pyplot as plt # pip install
import cv2  # pip install
import glob
from scipy.spatial import distance as dist  # pip install
```

```python
N = 50

x = np.random.rand(N) * 100.1
y = np.random.rand(N) * 100.1

k = 2
d = []
for i in range(N):
    d.append([x[i],y[i]])
perm=np.random.choice(range(N),k)
plt.figure()
plt.scatter(x, y)
plt.show()
```

```python

c = []
for x in perm:
    c.append(d[x])

print(c)
plt.figure()
for x in c:
    plt.scatter(x[0],x[1])
plt.scatter(100,100)
plt.show()
```

```python
max_iter=10
for it in range(max_iter):
    g = []
    for i in range(N):
        mn=1e9
        id=-1
        for j in range(k):
            me = dist.euclidean([c[j][0],c[j][1]], [d[i][0],d[i][1]])
            if (me < mn):
                mn = me
                id = j
        g.append(id)
    for j in range(k):
        sum = [0,0]
        cnt=0
        for i in range(N):
            if (g[i] == j):
                sum += d[i]
                cnt+= 1
        if not cnt:
            cnt=1
        sum[0] /=cnt
        sum[1] /= cnt
        c[j]=sum
    plt.figure()
    print(c[:][0],c[:][1])
    for x in c:
        plt.scatter(x[0],x[1])
    for x in d:
        plt.scatter(x[0],x[1])
    plt.show()
        
```
